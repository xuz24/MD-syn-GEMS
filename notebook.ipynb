```python
smiles = pd.read_csv('/Users/zijiexu/GEMS-LAB/jumpstart_bundle 2/data/synergy_labels/drug_chemical_info_with_smiles.csv')
smiles
```

```python
smiles['normalized_drugName'] = smiles['drugName']
smiles['normalized_drugName'] = (
    smiles['normalized_drugName']
    .str.lower()
    .str.strip()
)

smiles['normalized_drugNameOfficial'] = smiles['drugNameOfficial']
smiles['normalized_drugNameOfficial'] = (
    smiles['normalized_drugNameOfficial']
    .str.lower()
    .str.strip()
)

smiles
```

```python
smiles_drugName = smiles.copy().set_index('normalized_drugName')
smiles_drugNameOfficial = smiles.copy().set_index('normalized_drugNameOfficial')
```

```python
dbid_to_drug = pd.read_csv(CANONICAL_DATA_DIR + 'drugcomb_name_to_dbid_mapping_improved.csv').set_index('drugbank_id')
dbid_to_drug
```

```python
test_id = 'DB00544'
```

```python
def get_names(id):
    return dbid_to_drug['name_norm'].get(id)
```

```python
def get_smiles(names):
    for name in names:
        if name in smiles_drugName.index:
            print(f'{name} found in smiles_drugName')

            smiles = smiles_drugName.loc[name, 'smilesString']

            # handle duplicate indices
            if isinstance(smiles, pd.Series):
                smiles = smiles.iloc[0]

            return smiles

        elif name in smiles_drugNameOfficial.index:
            print(f'{name} found in smiles_drugNameOfficial')

            smiles = smiles_drugNameOfficial.loc[name, 'smilesString']

            # handle duplicate indices
            if isinstance(smiles, pd.Series):
                smiles = smiles.iloc[0]

            return smiles

    return None
```

```python
dbids = (
    set(fold_splits['drug1_dbid'].dropna().unique()) |
    set(fold_splits['drug2_dbid'].dropna().unique())
)
len(dbids)
```

```python
smiles_list = []
no_ids = 0
no_smiles = 0
no_smiles_dbids = []
for id in dbids:
    names = get_names(id)
    if names is None:
        no_ids += 1
        print(f'[NO NAME] id: {id}')
        continue
    
    smiles = get_smiles(names)
    if smiles is None or pd.isna(smiles):
        no_smiles += 1
        no_smiles_dbids.append(id)
        print(f'[NO SMILES] id: {id}, names: {names}')
        continue
    smiles_list.append(smiles)
    
```

```python
print(f"No ids: {no_ids}")
print(f"No smiles: {no_smiles}")
no_smiles_dbids
```
